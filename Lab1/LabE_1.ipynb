```python
from collections import Counter
import operator
```

# Q1)a)i)ii)iii) 

```python
file = open("f1.txt","r")
nlines = 0
nsentence = 0
nwords = 0
ncharacters = 0
for lines in file:
    nsentence +=  lines.count('. ')
    line = lines.strip("\n")
    words = line.split()
    nlines += 1
    nwords += len(words)
    for char in words:
        ncharacters += len(char)

file.close()
print("lines:", nlines, "Sentence:",nsentence, "words:", nwords, "characters:", ncharacters)
```

# Q1)b

```python
text_file = open('f1.txt', 'r')
text = text_file.read()

#cleaning
text = text.lower()
words = text.split()
words = [word.strip('.,!;()[]') for word in words]
words = [word.replace("'s", '') for word in words]

#finding unique
unique = []
for word in words:
    if word not in unique:
        unique.append(word)

#sort
unique.sort()

#print
print(len(unique))
print(unique)
```

# Q1)c)

```python
file = open("f1.txt","r")
content = file.read()
content_list = content.split(" ")
unique_words = set(unique)
for words in unique_words :
    print('Frequency of ', words , 'is :', content_list.count(words))
```

# Q2

```python
def opening(x):

    with open(x,'r') as y:
        file_contents = y.read()
        file_contents = file_contents.lower()
        words = file_contents.split()
        words = [word.strip('.,!;()[]') for word in words]
        words = [word.replace("'s", '') for word in words]

    counts = dict(Counter(words))
    return counts
```

```python
user_input = (input())
user_input = user_input.lower()

i_words = user_input.split()
i_words = [word.strip('.,!;()[]') for word in i_words]
i_words = [word.replace("'s", '') for word in i_words]

i_counts = dict(Counter(i_words))

final_list_f1 = set(opening("f1.txt"))&set(i_counts)

final_list_f2 = set(opening("f2.txt"))&set(i_counts)

final_list_f3 = set(opening("f3.txt"))&set(i_counts)

final_list_f4 = set(opening("f4.txt"))&set(i_counts)

final_list_f5 = set(opening("f5.txt"))&set(i_counts)

files_dict = {}
files_dict["f1"] = len(final_list_f1)
files_dict["f2"] = len(final_list_f2)
files_dict["f3"] = len(final_list_f3)
files_dict["f4"] = len(final_list_f4)
files_dict["f5"] = len(final_list_f5)

print(sorted(files_dict.items(), key=operator.itemgetter(1),reverse=True))
```
